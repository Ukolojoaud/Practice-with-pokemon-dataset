```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('pokemon.csv',index_col=0)
```

```python
df.head()
```

# Attack by defense using lmplot

```python
sns.lmplot(x='Attack', y='Defense', data=df)
plt.show()
```

```python
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg=False, hue='Stage')
plt.ylim(0, 200)
plt.xlim(0, 160)
plt.show()
```

Using boxplot

```python
plt.figure(figsize=(9, 6))
sns.boxplot(data=df)
plt.show()
```

To remove Total, Stage, legendary

```python
sta_df=df.drop(['Total','Stage','Legendary'], axis=1)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=sta_df)
plt.show()
```

VIOLIN PLOT (and also applying theme)

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,8))
sns.violinplot(x='Type 1', y='Attack', data=df,inner=None, palette=colortype)


sns.swarmplot(x='Type 1', y='Attack', data=df, color='k', alpha=0.7)
plt.title('Attack by Type')

plt.show()
```

using swarm plot

```python
plt.figure(figsize=(9, 6))
sns.swarmplot(x='Type 2', y='Attack', data=df, palette=colortype)
plt.show()
```
